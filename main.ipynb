```python
def make_choice(scene):
    options = scene['options']
    for i, option in enumerate(options):
        print(to_file(f"{i + 1}. {option}"))

    print(to_file("\nСделайте свой выбор (введите номер ответа):"))
    while True:
        try:
            user_choice_index = to_file(int(input()), is_input=True)
            if 1 <= user_choice_index <= len(options):
                user_choice = list(options)[user_choice_index - 1]
                print(to_file(f"Вы выбрали: {user_choice_index}.{user_choice}\n"))
                break
            else:
                print(to_file(f"Пожалуйста, выберите число от 1 до {len(options)}."))
        except ValueError:
            print(to_file("Пожалуйста, введите число."))

    return user_choice

def follow_witness_branch(user_choice, branch, correct_answer="Наследство"):
    while user_choice != correct_answer:
        print(branch[user_choice].replace(SPACE_SIZE, ''))
        print(to_file("Ответ неверный. Попробуйте еще раз\n"))
        user_choice = make_choice(scene)

    print(to_file("Ответ верный.\n"))


def follow_police_office_2_branch(correct_answer='кошелек'):
    print(to_file("\nВведите свой ответ словом:"))
    attempts = []
    while True:
        user_choice = to_file(input().lower(), is_input=True)
        if user_choice.isdigit():
            print(to_file("Введите свой ответ словом.\n"))
            continue
        if user_choice.capitalize() not in things:
            print(to_file("Выберите вещь из списка\n"))
            continue
        elif user_choice.capitalize() in attempts:
            print(to_file(f"Вы уже делали такой выбор. Попробуйте еще раз\nВы попробовали: {', '.join(attempts)}"))
            continue

        attempts.append(user_choice.capitalize())
        if user_choice != correct_answer:
            print(to_file(f"Ответ неверный. Попробуйте еще раз\nВы попробовали: {', '.join(attempts)}"))
        else:
            break

    print(to_file("Ответ верный.\n"))

def follow_explore_branch(user_choice):
    if user_choice == 'Поговорить с семьей убитого':
        people = {'Мать убитого': "Спектакль", 'Сын убитого': "Дом", 'Жена убитого': "Салон красоты"}
        print(to_file("Введите запись в формате одного слово, без предлога. Пример: Мать убитого - Спектакль"))
        notes_dict = {}
        attempts = set()
        while len(people) > 0:
            user_input = to_file(input(), is_input=True)
            if ' - ' in user_input and '  ' not in user_input:
                person, place = user_input.split(' - ')
            else:
                print(to_file("Введите свой ответ в правильном формате."))
                continue
            if person.isdigit() or place.isdigit():
                print(to_file("Введите свой ответ словом."))
                continue
            if person.lower() in attempts:
                    print(to_file(f"Вы уже записали этого человека."))
                    continue
            if person.lower().capitalize() not in people:
                print(to_file("Запишите человека из списка."))
                continue
            if place.lower().capitalize() != people[person.lower().capitalize()]:
                print(to_file("Введите правильное место."))
                continue

            notes_dict[person.lower()] = place.strip().lower()
            attempts.add(person.lower())
            people.pop(person.lower().capitalize())

            print(to_file(f"Сейчас блокнот выглядит так:"))
            print(to_file('\n'.join([f"{key.capitalize()} - {value.capitalize()}" for key, value in notes_dict.items()])))

            if people: print(to_file(f"Осталось сделать запись о: {', '.join(people)}\n"))


def to_file(text, is_input=False, file_name="file.txt"):
    with open(file_name, 'a') as file:
        file.write(str(text) + "\n") if is_input == False else file.write("input: " + str(text) + "\n")

    return text
```
